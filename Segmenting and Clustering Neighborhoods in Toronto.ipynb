### Importing the Dataframe

```python
import pandas as pd
url = 'https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&direction=prev&oldid=1012023397'
df=pd.read_html(url)
```

```python
maindf=df[0]
maindf
```

### Removing Rows where Borough is not assigned

```python
maindf.drop(maindf[maindf['Borough']=='Not assigned'].index, inplace=True)
maindf
```

### Setting Neighbourhood = Borough if Neighbourhood is not assigned

```python
maindf.loc[maindf['Neighbourhood']=='Not assigned', 'Borough'] = maindf['Borough']
maindf.reset_index(inplace=True, drop=True)
maindf
```

### Adding Markers

```python
markers = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(newdf.Latitude, newdf.Longitude):
    markers.add_child(
        folium.features.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
map.add_child(markers)
```
